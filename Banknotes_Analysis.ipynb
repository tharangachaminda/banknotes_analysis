# Banknotes Authentication
### What is Banknotes Authentication?
<div class="alert alert-block alert-success">
Banknote analysis, also known as banknote authentication or banknote verification, is a critical process employed in the field of financial security and fraud detection. As digital transactions become increasingly prevalent, ensuring the authenticity and integrity of physical currency remains crucial.
    
Banknote analysis refers to the examination of paper currency to determine its legitimacy and identify potential counterfeits. It involves utilizing advanced technologies, techniques, and expertise to assess various security features embedded in banknotes, such as watermarks, security threads, holograms, and special inks.
</div>

## Data Preparation

### Split data into training and testing sets

```python
X = df.drop('class', axis=1)
y = df['class']
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

<a id="model-building"></a>
## Model Building
### Logistic Regression (LR)
#### Hyperparameter Tuning with **GridSearchCV**

```python
lr = LogisticRegression()
lr.get_params()
```

```python
import warnings
warnings.filterwarnings('ignore')
# parameter grid
params_lr = {
    'C': np.logspace(-3,3,7),
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'lbfgs', 'saga']
}

grid_cv = GridSearchCV(estimator=lr, param_grid=params_lr, cv=10)
grid_cv.fit(X_train, y_train)
```

```python
print("Best params: ", grid_cv.best_params_)
print("Best score: ", grid_cv.best_score_)
```

```python
lr_bst = LogisticRegression(C=1.0, penalty='l1', solver='saga')
lr_bst.fit(X_train, y_train)
```

```python
lr_bst.score(X_test, y_test)
```

```python
model_accuracy = {}
model_accuracy = {'LR': [round(lr_bst.score(X_test, y_test)*100, 2)]}
```

```python
model_accuracy
```

```python
lr_predict = lr_bst.predict(X_test)
lr_predict
```

#### Classification Report (Logistic Regression)

```python
print(metrics.classification_report(y_test, lr_predict))
```

#### Confusion Matrix (Logistic Regression)

```python
cm_lr = metrics.confusion_matrix(y_test, lr_predict, labels=[1, 0])
cm_display_lr = metrics.ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=['Valid', 'Invalid'])
cm_display_lr.plot()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

<a id="random-forest"></a>
### Random Forest Classifier
#### Hyperparameter Tuning with **GridSearchCV()**

```python
rf = RandomForestClassifier()

rf.get_params()
```

```python
params_rf = {
    'n_estimators': [50, 100, 150],
    'criterion': ["gini", "entropy", "log_loss"],
    #'max_depth': [1, 2, 3],
    #'min_samples_split': [2, 3, 4],
    #'min_samples_leaf': [1, 2, 3]
}

grid_cv_rf = GridSearchCV(estimator=rf, param_grid=params_rf, cv=5)
grid_cv_rf.fit(X_train, y_train)
```

```python
print("Best params: ", grid_cv_rf.best_params_)
print("Accuracy score: ", grid_cv_rf.best_score_)
print("Best estimator: ", grid_cv_rf.best_estimator_)
```

```python
rf_bst = RandomForestClassifier()
rf_bst.fit(X_train, y_train)
rf_bst.score(X_test, y_test)
```

```python
model_accuracy |= {'RF': [round(rf_bst.score(X_test, y_test)*100, 2)]} # merge: this is how we add new values the a dict
model_accuracy
```

#### Classification Report (Random Forest)

```python
rf_predict = rf_bst.predict(X_test)
print(metrics.classification_report(y_test, rf_predict))
```

#### Confusion Matrix (Random Forest)

```python
cm_rf = metrics.confusion_matrix(y_test, rf_predict, labels=[1, 0])
cm_display_rf = metrics.ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=['Valid', 'Invalid'])
cm_display_rf.plot()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

#### Hyperparameter Tuning with **GridSearchCV()**

```python
svc = svm.SVC()

svc.get_params()
```
